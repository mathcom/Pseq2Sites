```python
import os
import tqdm
import pandas as pd
import numpy as np

from rdkit.Chem import MolFromSmiles, MolToSmiles, MolFromMol2File

from Bio import PDB
from Bio.PDB import Select, PDBIO
from Bio.PDB.PDBParser import PDBParser
```

# 1. Load complex list

```python
input_dir = os.path.join('data', 'pdbbind', '2011-2019')
```

```python
output_dir = os.path.join('data_preprocessed')
if not os.path.exists(output_dir):
    os.mkdir(output_dir)
    
output_dir = os.path.join(output_dir, 'pdbbind')
if not os.path.exists(output_dir):
    os.mkdir(output_dir)
```

```python
complex_list = os.listdir(input_dir)
```

# 2. Remove HEATM

```python
class NonHetSelect(Select):
    def accept_residue(self, residue):
        return 1 if residue.id[0] == " " else 0

def make_file_HEATM_removed(pdbid, input_dir, output_dir):
    filepath_src = os.path.join(input_dir, pdbid, f"{pdbid}_protein.pdb")
    filepath_tar = os.path.join(output_dir, f"{pdbid}_remove_HEATM_protein.pdb")
    
    pdb = PDBParser().get_structure(pdbid, filepath_src)
    
    io = PDBIO()
    io.set_structure(pdb)
    io.save(filepath_tar, NonHetSelect())
```

```python
for pdbid in tqdm.tqdm(complex_list):
    make_file_HEATM_removed(pdbid, input_dir, output_dir)
```

# 3. Load Binding sites info

```python
amino_acids_short = {
    "ALA":"A",
    "ARG":"R",
    "ASN":"N",
    "ASP":"D",
    "CYS":"C",
    "GLU":"E",
    "GLN":"Q",
    "GLY":"G",
    "HIS":"H",
    "ILE":"I",
    "LEU":"L",
    "LYS":"K",
    "MET":"M",
    "PHE":"F",
    "PRO":"P",
    "SER":"S",
    "THR":"T",
    "TRP":"W",
    "TYR":"Y",
    "VAL":"V",
    "SEC":"U",
    "PYL":"O"
}
```

```python
def get_info(pdbid, input_dir, pocket_dir):
    ## init parser
    pdb_parser = PDB.PDBParser(QUIET = True)
    
    ## protein info.
    chain_name_list = []
    pdb_sequence_list = []
    seq_length_list = []
    protein_atom_coords = []
    protein_atom_residue_list = []
    reindex = 0
    
    ## data load
    structure = pdb_parser.get_structure(pdbid, os.path.join(input_dir, f'{pdbid}_remove_HEATM_protein.pdb'))
    
    for chain_name in structure[0].child_dict.keys():
        chain = structure[0][chain_name]
        
        pdb_sequence = ""
        for residue in chain.get_residues():
            if residue.resname in amino_acids_short:
                pdb_sequence += amino_acids_short[residue.resname]
                
                for atom in residue:
                    protein_atom_coords.append(atom.get_coord())
                    protein_atom_residue_list.append(reindex)
                
                reindex += 1
                
        if len(pdb_sequence) > 0:
            chain_name_list.append(chain_name)
            pdb_sequence_list.append(pdb_sequence)
            seq_length_list.append(len(pdb_sequence))
            
    ## pocket info.
    pocket_coords = []
    
    ## data load
    pocket_structure = pdb_parser.get_structure(pdbid, os.path.join(pocket_dir, pdbid, f'{pdbid}_pocket.pdb'))
    
    for chain_name in pocket_structure[0].child_dict.keys():
        chain = pocket_structure[0][chain_name]
        for residue in chain.get_residues():
            if residue.resname in amino_acids_short:
                for atom in residue:
                    pocket_coords.append(atom.get_coord())
                    
    ## matching pocket info.
    binding_index = []
    protein_atom_coords = np.array(protein_atom_coords)
    pocket_coords = np.array(pocket_coords)
    
    for i in range(len(pocket_coords)):
        tmp_coords = pocket_coords[i,:3]
        ind = np.where((protein_atom_coords == tmp_coords).all(axis=1))[0][0]
        binding_index.append(protein_atom_residue_list[ind])
        
    binding_index = sorted(list(set(binding_index)))
    binding_index = [str(x) for x in binding_index]
    
    total_seq_length = np.sum(np.array(seq_length_list))
    seq_length_list = [str(x) for x in seq_length_list]
    
    return {
        "PDB": pdbid,
        "Chain": ",".join(chain_name_list),
        "Sequence": ",".join(pdb_sequence_list),
        "Total_seq_length": total_seq_length,
        "Chain_seq_length": ",".join(seq_length_list),
        "BS": ",".join(binding_index),
    }
```

```python
info_results = []
for pdbid in tqdm.tqdm(complex_list):
    info_results.append(get_info(pdbid, output_dir, input_dir))
```

```python
data_df = pd.DataFrame(info_results)
```

# 4. Remove complex over than 1,500 protein seq length

```python
data_df = data_df.query("Total_seq_length <= 1500").reset_index(drop=True)
data_df
```

# 5. Remove complex over than 160 SMILES length

```python
def get_smiles(pdbid, input_dir):
    filepath = os.path.join(input_dir, pdbid, f"{pdbid}_ligand.mol2")
    mol = MolFromMol2File(filepath)
    
    if mol:
        return MolToSmiles(mol, isomericSmiles=False, kekuleSmiles=True)
```

```python
for i, pdbid in enumerate(data_df['PDB']):
    try:
        data_df.loc[i, "SMILES"] = get_smiles(pdbid, input_dir)
        data_df.loc[i, "SMILES_length"] = len(data_df.loc[i, "SMILES"])
    except TypeError:
        data_df.loc[i, "SMILES"] = None
        data_df.loc[i, "SMILES_length"] = 999
```

```python
data_df = data_df.query("SMILES_length <= 160").reset_index(drop=True)
```
