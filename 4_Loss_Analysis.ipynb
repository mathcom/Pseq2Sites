```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
input_dir = os.path.join("outputs", "pdbbind", "CV")
```

```python
frames = []
for i in range(5):
    filepath = os.path.join(input_dir, f"fold{i}", "history.csv")
    df = pd.read_csv(filepath)
    df.loc[:,"Fold"] = f"Fold {i}"
    frames.append(df)
```

```python
df = pd.concat(frames).reset_index(drop=True)
```

```python
sns.set_theme(style="whitegrid")
```

```python
sns.lineplot(data=df, x="Epoch", y="Loss_tr", hue="Fold")
```
